```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.tree import DecisionTreeRegressor
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

```python
df3=pd.read_csv("C:\\Users\\bkprlglu\\OneDrive - The University of Memphis\\desktop\\CSE 6242 Project\\df3.csv")
```

```python
df3=df3.drop(["Unnamed: 0"],axis=1)
```

```python
df3['population_density'].corr(df3['pollution_death'])
```

```python
df3.corr()
```

```python
train1,test1 = train_test_split(df3, test_size=0.2, random_state=42)
```

```python
X1_train = train1[['population_density','Population','NO2_tenyears','SO2_tenyears']]
y1_train = train1['resp_death']
X1_test = test1[['population_density','Population','NO2_tenyears','SO2_tenyears']]
y1_test = test1['resp_death']
```

```python
p_model = LinearRegression()
p_model.fit(X1_train, y1_train)

predicted_no2 = p_model.predict(X1_test)

mean_squared_error(y1_test, predicted_no2)


```

```python
mean_absolute_error(y1_test, predicted_no2)
```

```python
mean_absolute_percentage_error(y1_test, predicted_no2)
```

```python
p_model1 = DecisionTreeRegressor(max_depth=None,random_state=7)
p_model1.fit(X1_train, y1_train)

predicted_no2 = p_model1.predict(X1_test)

mean_squared_error(y1_test, predicted_no2)
```

```python
mean_absolute_error(y1_test, predicted_no2)
```

```python
mean_absolute_percentage_error(y1_test, predicted_no2)
```

```python
p_model11 = RandomForestRegressor(n_estimators=100,
                                  random_state=15)

p_model11.fit(X1_train, y1_train)

predicted_no22 = p_model11.predict(X1_test)

mean_squared_error(y1_test, predicted_no22)
```

```python
mean_absolute_error(y1_test, predicted_no22)
```

```python
mean_absolute_percentage_error(y1_test, predicted_no22)
```
